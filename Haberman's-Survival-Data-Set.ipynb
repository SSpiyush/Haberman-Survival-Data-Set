```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('haberman.csv')
df
```

```python
i = df.rename(columns={'30':'AGE', '64':'YEAR', '1':'AXILLARY NODES DETECTED', '1.1':'SURVIVAL STATUS'})
i
```

```python
i['SURVIVAL STATUS'].replace({1:'survived above 5yrs', 2:'died within 5yrs'}, inplace=True)
i
```

```python
i['SURVIVAL STATUS'].value_counts()
```

# 2D Scatter Plot

```python
sns.set_style('darkgrid');
sns.FacetGrid(i, hue='SURVIVAL STATUS', size=6) \
   .map(plt.scatter, "AGE", "AXILLARY NODES DETECTED") \
   .add_legend();
plt.title('PLOT BETWEEN AXILLARY NODES DETECTED Vs AGE')
plt.show()
```

# Observation:
1) In this 2D scatter plot we can conclude that the patient younger than 32 years have the maximum chance to survive for 5 years and more.

```python
sns.set_style('darkgrid');
sns.FacetGrid(i, hue='SURVIVAL STATUS', size=6) \
   .map(plt.scatter, "YEAR", "AXILLARY NODES DETECTED") \
   .add_legend();
plt.title('PLOT BETWEEN AXILLARY NODES DETECTED Vs YEAR')
plt.show() 
```

## Observation:
1) we cannot conclude anything, considering year and axillary nodes detected in 2-D scatter plot.



# 3D Scatter Plot

```python
plt.close();
sns.set_style('darkgrid');
sns.pairplot(i, hue='SURVIVAL STATUS', size=4);
plt.show()
```

# Histogram,PDF,CDF

```python
import numpy as np
i_survived_above_5yrs = i.loc[i['SURVIVAL STATUS'] == 'survived above 5yrs' ];
i_died_within_5yrs = i.loc[i['SURVIVAL STATUS'] == 'died within 5yrs'];

plt.plot(i_survived_above_5yrs['AXILLARY NODES DETECTED'], np.zeros_like(i_survived_above_5yrs['AXILLARY NODES DETECTED']), 'o',label='survived above 5yrs')
plt.plot(i_died_within_5yrs['AXILLARY NODES DETECTED'], np.zeros_like(i_died_within_5yrs['AXILLARY NODES DETECTED']), 'o',label='died within 5yrs')
plt.xlabel('AXILLARY NODES DETECTED')
plt.title('PLOT ON AXILLARY NODES DETECTED')
plt.legend()
plt.show()
```

```python
sns.FacetGrid(i, hue='SURVIVAL STATUS', size=6) \
   .map(sns.distplot, 'AXILLARY NODES DETECTED') \
   .add_legend();
plt.title('PDF AND HISTOGRAM BASED ON AXILLARY NODES DETECTED ')
plt.show()

```

# Observation:
1) In this histogram we can conclude that a person who is detected with less than 2 axillary nodes have greater tendency to live 
   for more than 5 years.

```python
sns.FacetGrid(i, hue='SURVIVAL STATUS', size=6) \
   .map(sns.distplot, 'AGE') \
   .add_legend();
plt.title('PDF AND HISTOGRAM BASED ON AGE ')
plt.show()
```

```python
sns.FacetGrid(i, hue='SURVIVAL STATUS', size=6) \
   .map(sns.distplot, 'YEAR') \
   .add_legend();
plt.title('PDF AND HISTOGRAM BASED ON YEAR ')
plt.show()
```

```python
counts, bin_edges = np.histogram(i_survived_above_5yrs['AXILLARY NODES DETECTED'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='SURVIVED ABOVE 5YRS(PDF)')
plt.plot(bin_edges[1:], cdf,label='SURVIVED ABOVE 5YRS(CDF)')
plt.xlabel('AXILLARY NODES DETECTED')
plt.ylabel('PDF/CDF')
plt.title('PDF AND CDF BASED ON AXILLARY NODES DETECTED')
plt.legend()
plt.show()
```

```python
counts, bin_edges = np.histogram(i_survived_above_5yrs['AXILLARY NODES DETECTED'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='SURVIVED ABOVE 5YRS(PDF)')
plt.plot(bin_edges[1:], cdf,label='SURVIVED ABOVE 5YRS(CDF)')


counts, bin_edges = np.histogram(i_died_within_5yrs['AXILLARY NODES DETECTED'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='DIED WITHIN 5YRS(PDF)')
plt.plot(bin_edges[1:], cdf,label='DIED WITHIN 5YRS(CDF)')
plt.xlabel('AXILLARY NODES DETECTED')
plt.ylabel('PDF/CDF')
plt.title('PDF AND CDF BASED ON AXILLARY NODES DETECTED')
plt.legend()
plt.show();
```

# Mean, Variance And Standard-Deviation

```python
print("Means:")
print(np.mean(i_survived_above_5yrs['AXILLARY NODES DETECTED']))
print(np.mean(i_died_within_5yrs['AXILLARY NODES DETECTED']))

print("\nStd-dev:");
print(np.std(i_survived_above_5yrs['AXILLARY NODES DETECTED']))
print(np.std(i_died_within_5yrs['AXILLARY NODES DETECTED']))


```

# Median, Percentile, Quantile, IQR, MAD

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(i_survived_above_5yrs['AXILLARY NODES DETECTED']))
#Median with an outlier
print(np.median(i_died_within_5yrs['AXILLARY NODES DETECTED']))

print("\nQuantiles:")
print(np.percentile(i_survived_above_5yrs['AXILLARY NODES DETECTED'],np.arange(0, 100, 25)))
print(np.percentile(i_died_within_5yrs['AXILLARY NODES DETECTED'],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(i_survived_above_5yrs['AXILLARY NODES DETECTED'],90))
print(np.percentile(i_died_within_5yrs['AXILLARY NODES DETECTED'],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(i_survived_above_5yrs['AXILLARY NODES DETECTED']))
print(robust.mad(i_died_within_5yrs['AXILLARY NODES DETECTED']))

```
